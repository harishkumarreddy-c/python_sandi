```python
xMatrix = [
    [1,2,3],
    [1,2,3],
    [1,2,3],
]

for row in xMatrix:
    for col in row:
        print(col)

print(xMatrix[0])

r = 3
# if xMatrix[r]:
if len(xMatrix) > r:
    for i in xMatrix[r]:
        print(i)
    else:
        print("Range is completed.")
else:
    print("Range is out of bound.")
```

```python
arr = [1, 2, "three", "5", 6.75 ]
v_arr = [] # get true where a number/float is present and false where a string is present
for item in arr:
    if type(item) is str:
        v_arr.append(False)
    else:
        v_arr.append(True)
print(arr)
print(v_arr)
```

```python
v_arr2 = [x for x in arr if type(x) is str]
print(arr)
print(v_arr2)
```

```python
v_arr3 = [True if x else False for x in arr if type(x) is str]
print(arr)
print(v_arr3)
```
