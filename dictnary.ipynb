```python
# simple Dictionary
mylist = ["Alice", 30, "New York", ["Python", "Data Science", "Machine Learning"], 95.5]

myDict = {
    "name": "Alice", 
    "age": 30, 
    "city": "New York",
    "skills": ["Python", "Data Science", "Machine Learning"],
    "score": 95.5
    }
print(myDict)
print(type(myDict))
```

```python
print(mylist[0])  
print(myDict["name"])
print(myDict["skills"])

# print discriptive message
print(f"{myDict['name']} is {myDict['age']} years old and lives in {myDict['city']}.")
```

```python
# getter and setter methods

print(myDict.get("name"))
myDict["name"] = "Bob"
print(myDict.get("name"))
```

```python
print(myDict["skills"])
```

```python
print(myDict.get("skills")[2])
print(mylist[3][2])
```

```python
# print(myDict["dob"])
print(myDict.get("dob", "Not Found"))
```
