```python
from datetime import datetime

dob = "1998-01-01"
print("Date of Birth:", dob)
print("Type of Date of Birth:", type(dob))

# Convert string to datetime object
# dob_datetime = datetime.strptime(dob, "%Y-%m-%d")
dob = datetime.strptime(dob, "%Y-%m-%d") # epoch time
print("Date of Birth (datetime):", dob)
print("Type of Date of Birth (datetime):", type(dob))
```
