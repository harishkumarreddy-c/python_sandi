```python
myChar = 'a'

print(myChar)
print(type(myChar))
print(myChar.upper())
```

```python
# string operations

myString = "hello world"
print(myString)
print(type(myString))

# case conversion
print(myString.lower())
print(myString.upper())
print(myString.title())

# size of string
print(len(myString))

```

```python
# indexing and slicing
print(myString[5])  
print(myString[4:5])  
# print(myString[len(myString)-1]) 
print(myString[-1])
print(myString[-2])
print(myString[-5:-1])
print(myString[4:])
print(myString[:5])
print(myString[-5:])
print(myString[:-5])
print(myString[:])

```

```python
# skipping characters
print(myString[::2])
print(myString[1:8:2])
```

```python
# casting
myInt = 123
print(myInt)
print(type(myInt))

f = float(myInt)
print(f)
print(type(f))

myString = str(myInt)
print(myString)
print(type(myString))

age = "10"
print(age)
print(type(age))
age = int(age)
print(age)
print(type(age))
```

```python
str = "123.45j"
print(str)
print(type(str))
str = complex(str)  # This will raise a ValueError since "abc" cannot be converted to an integer
print(str)
print(type(str))
```

```python
# concatenation
a = "hello"
b = "world"
c = a + b
d = 123

print(c)
print(c + str(d))
```

```python
# formatting
name = "Alice"
age = 30
print("My name is {} and I am {} years old.".format(name, age))
print(f"My name is {name} and I am {age} years old.")
```

```python
msg = "My name is Alice and I am 30 years old."
print(msg)
print(type(msg))

charList = list(msg)
print(charList)
print(type(charList))

worsList = msg.split()
print(worsList)
print(type(worsList))
```

```python
newMsg = "jxy".join(charList)
print(newMsg)

encVal = msg.encode('utf-8')
print(encVal.decode('utf-8'))


```

```python
# base64 encoding
import base64

encoded = base64.b64encode(msg.encode('utf-8'))
print(encoded)
print(encoded.decode('utf-8'))
print(base64.b64decode(encoded).decode('utf-8'))


```

```python
# md5 hashing
import hashlib

hashObj = hashlib.md5(msg.encode('utf-8'))
print(hashObj.hexdigest())
```
