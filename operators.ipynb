### Assignment Operators:

```python
# Assignment Operators

x = 5
y=3

sum = x + y
print("Sum:", sum)

difference = x - y
print("Difference:", difference)

mutiply = x * y
print("Product:", mutiply)

div = x / y
print("Quotient:", div)

power = x ** y
print("Power:", power)


# ---
fdiv = x // y
print("Floor Division:", fdiv)

mod = x % y
print("Modulus:", mod)


```

```python
# find odd or even number
num = 3
res = num % 2
print("Remainder:", res)
if res == 0:
    print("Even")
else:
    print("Odd")
```

```python
x=0
print(x)
x += 5 # x = x + 5
print(x)

x -= 2 # x = x - 2
print(x)

x *= 3 # x = x * 3
print(x)

x /= 4 # x = x / 4
print(x)

x **= 2 # x = x ** 2
print(x)

x //= 3 # x = x // 3
print(x)

x %= 4 # x = x % 4
print(x)
```

```python
x = 5
y = 3
z = 2

print(x==y) # False
print(x!=y) # True
print(x>y) # True
print(x<y) # False
print(x>=y) # True
print(x<=y) # False

print(x>y and x<z) # False
print(x>y or x<z) # True
print(not x>y) # False

'''
    true and true = true
    true and false = false
    false and true = false
    false and false = false

    true or true = true
    true or false = true
    false or true = true
    false or false = false

    not true = false
    not false = true
'''
```
