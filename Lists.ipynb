```python
# Simple List / Array
my_list = [1, 2,"three", 4.0, True]
print(my_list)
print(type(my_list))
print(len(my_list))
```

```python
print(my_list[0])
print(my_list[2])
print(my_list[4])
print(my_list[-1])
print(my_list[2:4])
print(my_list[-4:-1])
print(my_list[:3])
print(my_list[3:])
print(my_list[::2])
```

```python
my_list = [1, 2,"three", 4.0, True]
print(my_list)

my_list[1] = "two"
print(my_list)

# add new item to the end of the list
my_list.append("new item")
print(my_list)

# Add a new item at the first position / specified position
my_list.insert(0, "first item")
my_list.insert(2, "third item")
print(my_list)
```

```python
res = my_list.reverse()
print(my_list)
print(res)
```

```python
pop_val = my_list.pop()
slice_val = my_list[2:4]
print(pop_val)
print(slice_val)
```
